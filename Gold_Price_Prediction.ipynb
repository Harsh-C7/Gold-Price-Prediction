```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
df = pd.read_csv('/content/gld_price_data.csv')
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Date'], axis=1)
```

```python
df.head()
```

```python
correlation = df.corr()

plt.figure(figsize=(8, 8))
sns.heatmap(correlation, annot=True, cbar=True, cmap="Blues", fmt=".1f", annot_kws={"size":8}, square=True)
plt.show()
```

```python
plt.figure(figsize=(4, 3))
sns.displot(df['GLD'], color="green")
plt.show()
```

```python
x = df.drop(['GLD'], axis=1)
y = df['GLD']
print(x)
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
rf = RandomForestRegressor(n_estimators=100)
rf.fit(x_train, y_train)
```

```python
print("The Training R Square mean ", r2_score(y_train, rf.predict(x_train)))
```

```python
print("The Testing R Square mean ", r2_score(y_test, rf.predict(x_test)))
```
